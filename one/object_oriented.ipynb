```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Rango de 0 hasta 5, Numero de valores 11 / np.linspace
x=np.linspace(0,5,11)
y=x**2
```

```python
#Representa el todo y los axes son las graficas dentro del lienzo
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes_2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes.plot(x,y,'b')
axes_2.plot(x,y,'r')
axes_2.set_facecolor('gray')
fig.show()
```

```python
plt.hist(x, color='red', alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma con Color Personalizado')
plt.show()
```

```python
plt.pie(y)
plt.show()
```

```python
plt.scatter(x,y)  #correlacion entre la variable x e y
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
