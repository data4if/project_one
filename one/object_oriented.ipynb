```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Rango de 0 hasta 5, Numero de valores 11 / np.linspace
x=np.linspace(0,5,11)
y=x**2
# 
plt.plot(x,y,'bD-')
plt.show()
```

```python
plt.hist(x, color='red', alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma con Color Personalizado')
plt.show()
```

```python
plt.pie(y)
plt.show()
```

```python
plt.scatter(x,y)  #correlacion entre la variable x e y
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
