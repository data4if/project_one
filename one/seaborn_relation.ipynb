```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
markers = {'Lunch':'D','Dinner':'s'}
sns.scatterplot(data=tips,
                x='total_bill',
                y='tip', 
                hue='day', 
                style='time', 
                markers=markers, 
                size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,
                x='total_bill',
                y='tip')
```

```python
sns.relplot(data=tips,
                x='total_bill',
                y='tip', 
                hue='day', 
                style='time', 
                markers=markers, 
                size='size',
                #kind='line',
                col='time')
```
