```python
tips = sns.load_dataset('tips')
sns.displot(data=tips, x='total_bill')
```

```python
tips = sns.load_dataset('tips')
#KDE es distribucion por frecuencia
# sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
sns.displot(data=tips, 
            x='tip', 
            hue='sex', 
            kind='kde', 
            legend=True)
plt.show()
```

```python
# bins numero de barras para la frecuencia
sns.histplot(data=tips, 
             x='tip', 
             cumulative=False, 
             hue='sex', 
             stat='probability', #percentage #density #count
             multiple='dodge'); #dogde #stack 
```

```python
sns.kdeplot(data=tips,
            x='tip',
            hue='sex',
            fill=True);
```

```python
sns.ecdfplot(data=tips, x='tip',hue='sex');
plt.show()
```

```python
sns.displot(data=tips, x='tip',hue='sex', kind='kde');
```

```python
# Graficar variables categoricas
sns.countplot(data=tips,x='day',hue='sex');
```

```python
sns.histplot(data=tips, 
                y='sex', 
                x='tip', 
                hue='sex',
                multiple='fill')
                #dodge=True)
#tips.head()
```

```python
sns.kdeplot(data=tips, 
            x='tip', 
            hue='sex', 
            cumulative=False, 
            shade=True,
            bw_adjust=True,
            )
```

```python
# para trabajar graficos de distrubucion - Favorito
sns.displot(data=tips,
            x='tip',
            hue='sex', 
            kind='kde',
            multiple='stack')
```

```python
# Graficar variables categoricas
plt.figure(figsize=(10,10));
# Jugar con kaggle
sns.stripplot(data=tips, x='day',y='tip', hue='sex', dodge=True, color='0', marker="D");
sns.boxplot(data=tips, x='day',y='tip', hue='sex', dodge=True);
#              , bw_adjust=True,)
# plt.legend( title = 'Gender',  fontsixe=10, title_fontsize =14)
plt.show();
```

```python
sns.violinplot(data=tips, x='day',y='tip', hue='sex', dodge=True, palattte='pastel',split=True);
```

```python
# Grafica variables categoricas
sns.catplot(data=tips,x='day',y='tip', hue='smoker', dodge=True, kind='box',col='time')
```
